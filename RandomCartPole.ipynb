```python
import numpy as np
import pandas as pd
import random
```

```python
import gym
env = gym.make('CartPole-v0')
```

```python
env.seed(100) 
env.action_space.seed(100)
random.seed(100)
np.random.seed(100)
```

```python
def run_one_episode(env):
  state = env.reset()
  for step in range(200):
    a = random.randint(0,1)
    state, reward, done, info = env.step(a)
    if done:
      break
    print(state, reward, info)
  return step
```

```python
nosteps = run_one_episode(env)
if nosteps >= 200:
  print(f'SUCCESS')
else:
  print(f'FAIL')
print(f'nosteps = {nosteps}')
```
